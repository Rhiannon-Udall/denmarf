```python
import numpy as np
import corner
```

```python
import matplotlib
%matplotlib inline
%config InlineBackend.figure_format = "retina"
```

```python
samples = np.random.multivariate_normal([0, 0], [[1,0],[0,1]], size=10000)
```

```python
fig = corner.corner(samples, color="#C8102E", alpha=0.3, range=[(-14/4, 23/4), (-14/4, 14/4)], truths=[0,0], truth_color="white")
ax = fig.axes
ax[2].set_facecolor("#C8102E")
ax[0].remove(); ax[1].remove(); ax[3].remove()
xax = ax[2].axes.get_xaxis()
yax = ax[2].axes.get_yaxis()
xax.set_visible(False)
yax.set_visible(False)
ax[2].spines['top'].set_visible(False)
ax[2].spines['right'].set_visible(False)
ax[2].spines['bottom'].set_visible(False)
ax[2].spines['left'].set_visible(False)
ax[2].get_lines()[1].set_linewidth(10)
ax[2].get_lines()[2].set_linewidth(10)

fig.savefig("denmarf_logo.pdf", bbox_inches='tight', dpi=150)
```
